```python
"""A simple example of creating a figure with text rendered in LaTeX."""
#from thesis import COLOR_SECONDARY
import thesis
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("./thesis.mplstyle")
```

```python
# Initialise figure instance
fig, ax = plt.subplots(1, 1, figsize=thesis.set_size(), dpi=150)

# Plot
x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.sin(x+0.5*np.pi))
ax.plot(x, np.sin(x+1.0*np.pi))
ax.plot(x, np.sin(x+1.5*np.pi))

ax.set_xlim(0, 2 * np.pi)
ax.set_title("Sineusoids") #, color=COLOR_SECONDARY)
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$\sin (\theta)$')
```

```python
fig.savefig('example_1.pdf', format='pdf', bbox_inches='tight')
```
